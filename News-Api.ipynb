```python
import requests
import tkinter as tk
from tkinter import scrolledtext, messagebox

# Function to fetch news with pagination support
def fetch_news(api_key, query=None, category=None, sort_by='publishedAt', page=1):
    base_url = "https://newsapi.org/v2/top-headlines?"
    params = {
        'apiKey': api_key,
        'q': query,
        'category': category,
        'sortBy': sort_by,
        'language': 'en',
        'page': page,
        'pageSize': 5  # Number of articles per page
    }
    response = requests.get(base_url, params=params)
    return response.json()

# Function to display news articles in the text widget
def display_news(news_data):
    if news_data['status'] == 'ok':
        articles = news_data.get('articles', [])
        news_display.delete(1.0, tk.END)
        for i, article in enumerate(articles):
            news_display.insert(tk.END, f"{i + 1}. {article['title']}\n")
            news_display.insert(tk.END, f"   Source: {article['source']['name']}\n")
            news_display.insert(tk.END, f"   Author: {article.get('author', 'Unknown')}\n")
            news_display.insert(tk.END, f"   Published At: {article['publishedAt']}\n")
            news_display.insert(tk.END, f"   URL: {article['url']}\n\n")
    else:
        messagebox.showerror("Error", "Error fetching news articles.")

# Function to fetch and display news based on user inputs
def search_news():
    global current_page
    current_page = 1
    query = query_entry.get()
    category = category_entry.get()
    sort_by = sort_by_entry.get()
    news_data = fetch_news(api_key, query, category, sort_by, current_page)
    display_news(news_data)

# Function to go to the next page
def next_page():
    global current_page
    current_page += 1
    query = query_entry.get()
    category = category_entry.get()
    sort_by = sort_by_entry.get()
    news_data = fetch_news(api_key, query, category, sort_by, current_page)
    if news_data['articles']:
        display_news(news_data)
    else:
        messagebox.showinfo("Info", "No more articles available.")
        current_page -= 1

# Function to go to the previous page
def previous_page():
    global current_page
    if current_page > 1:
        current_page -= 1
        query = query_entry.get()
        category = category_entry.get()
        sort_by = sort_by_entry.get()
        news_data = fetch_news(api_key, query, category, sort_by, current_page)
        display_news(news_data)
    else:
        messagebox.showinfo("Info", "You are on the first page.")

# Replace with your actual API key
api_key = "5c9b4ce48ebb46eea9b3bcdc0fec1dd2"
current_page = 1

# Set up the GUI
root = tk.Tk()
root.title("News Aggregator")

tk.Label(root, text="Keyword:").grid(row=0, column=0, padx=10, pady=10)
query_entry = tk.Entry(root)
query_entry.grid(row=0, column=1, padx=10, pady=10)

tk.Label(root, text="Category:").grid(row=1, column=0, padx=10, pady=10)
category_entry = tk.Entry(root)
category_entry.grid(row=1, column=1, padx=10, pady=10)

tk.Label(root, text="Sort By:").grid(row=2, column=0, padx=10, pady=10)
sort_by_entry = tk.Entry(root)
sort_by_entry.grid(row=2, column=1, padx=10, pady=10)

search_button = tk.Button(root, text="Search", command=search_news)
search_button.grid(row=3, column=0, columnspan=2, padx=10, pady=10)

news_display = scrolledtext.ScrolledText(root, wrap=tk.WORD, width=80, height=20)
news_display.grid(row=4, column=0, columnspan=2, padx=10, pady=10)

prev_button = tk.Button(root, text="Previous Page", command=previous_page)
prev_button.grid(row=5, column=0, padx=10, pady=10, sticky='w')

next_button = tk.Button(root, text="Next Page", command=next_page)
next_button.grid(row=5, column=1, padx=10, pady=10, sticky='e')

root.mainloop()
```
